```python
import pandas as pd
from get_SAB import *
import copy
import numpy as np 
from pandas import DataFrame
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python

info = {}
info2 = {}
maxSAB = 0.0
with open('leaprRun/tape24','r') as f:
    lines = [line for line in f.readlines() if line[71:75] == '7  4']

    nbeta  = getVal(lines[4],5,'int')
    nalpha = getVal(lines[6],5,'int')
    numAlphaLines = ceil(nalpha/3)
    lines = lines[6:]

    for ibeta in range(nbeta):
        offset = 2
        beta = getVal(lines[0],1)
        alphas, sab = [], []
        for i in range(numAlphaLines):
            for j in range(3):
                alphas.append(getVal(lines[offset+i],2*j  ))
                sab.append(   getVal(lines[offset+i],2*j+1))
                if sab[-1] > maxSAB:
                    maxSAB = sab[-1]
        info[beta] = [alphas,sab]
        info2[beta] = sab
        lines = lines[numAlphaLines+offset:]

    betas = (list(info.keys()))
```

```python

for ibeta in [1,3,5,10,50,100]:
    beta = betas[ibeta]
    alphas = info[beta][0]
    sab    = info[beta][1]
    plt.plot(alphas,sab)
plt.yscale('log')
```

```python


sab = [info[beta][1][10] for beta in betas]
plt.plot(betas,sab)
sab = [info[beta][1][50] for beta in betas]
plt.plot(betas,sab)
sab = [info[beta][1][100] for beta in betas]
plt.plot(betas,sab)



#plt.plot(alphas,sab)
plt.yscale('log')
plt.show()

```

```python
info3 = copy.deepcopy(info2)
# info3['alphas'] = alphas
df = pd.DataFrame(data=info2)

plt.pcolor(df, norm=LogNorm(vmin=1e-20, vmax=maxSAB))#, cmap='PuBu_r')
plt.colorbar()

#plt.yticks(np.arange(0.5, len(df.index), 1), df.index)
#plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns)
plt.show()
```

```python
U, sigma, V = np.linalg.svd(df)
# V_df = pd.DataFrame(V)
# print(V_df)
print(len(sigma))
norms = []
numEigenVec = list(range(1,50,2))#+[50,100,150]
for numEigenvals in numEigenVec:
    A_approx = np.matrix(U[:, :numEigenvals]) * np.diag(sigma[:numEigenvals]) * np.matrix(V[:numEigenvals, :])
    A_approx_df = pd.DataFrame(A_approx,columns=betas)
    norms.append(np.linalg.norm(A_approx_df))
    plt.pcolor(A_approx_df, norm=LogNorm(vmin=1e-20, vmax=maxSAB))#, cmap='PuBu_r')
    plt.colorbar()
    plt.title(str(numEigenvals))
    plt.show()
```

```python
plt.plot(numEigenVec,norms)
plt.plot([numEigenVec[0],numEigenVec[-1]],[np.linalg.norm(df),np.linalg.norm(df)],'r')
plt.ylim([774000,774200])
```
